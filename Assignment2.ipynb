#Part 1:
Calculate the Mean, Median, Mode and Standard deviation.

```
import csv, numpy
from scipy import stats

def mean(arr):
  print(numpy.mean(arr))

def median(arr):
  print(numpy.median(arr))

def mode(arr):
  print(stats.mode(arr)[0])

def stddev(arr):
  print(numpy.std(arr))

def row_wise(_scores_, op):
  i = 1
  for row in _scores_:
    print("Row "+str(i)+": ", end=" ")
    op(row)
    i += 1

def column_wise(_scores_, op):
  for i in range(0, _scores_.shape[1]):
    print("Column "+str(i+1)+": ", end=" ")
    op(_scores_[:, i])

def overall(_scores_, op):
  op(_scores_.flatten())

with open('wine.csv') as csvfile:
  data = csv.reader(csvfile, delimiter=',')
  scores = []
  for row in data:
    try:
      temp_scores = list(map(float, row))
      scores.append(temp_scores)
    except ValueError:
      pass

  scores = numpy.array(scores) 

  print("Row-wise mean of given dataset: ")
  row_wise(scores, mean)
  print("\nRow-wise median of given dataset: ")
  row_wise(scores, median)
  print("\nRow-wise mode of given dataset: ")
  row_wise(scores, mode)
  print("\nRow-wise std dev of given dataset: ")
  row_wise(scores, stddev)

  print("\nColumn-wise mean of given dataset: ")
  column_wise(scores, mean)
  print("\nColumn-wise median of given dataset: ")
  column_wise(scores, median)
  print("\nColumn-wise mode of given dataset: ")
  column_wise(scores, mode)
  print("\nColumn-wise std dev of given dataset: ")
  column_wise(scores, stddev)

  print("\nOverall mean of given dataset: ")
  overall(scores, mean)
  print("\nOverall median of given dataset: ")
  overall(scores, median)
  print("\nOverall mode of given dataset: ")
  overall(scores, mode)
  print("\nOverall std dev of given dataset: ")
  overall(scores, stddev)
```

#Part 2:
Calculate the pearson correlation and covariance matrix.

```
import pandas as pd

data = pd.read_csv('wine.csv')
corr_mat = data[['Alcohol', 'Proline']].corr(method="pearson")
print("Pearson correlation coefficient of Alcohol and proline feature is: "+str(corr_mat.at['Alcohol', 'Proline']))

feature5 = data[data.columns[0:5]]
print("\nCovariance matix of first five features:- ")
feature5.cov()
```
